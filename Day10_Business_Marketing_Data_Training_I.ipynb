```python
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
%matplotlib inline
```

## Obtaining Data

```python
df = pd.read_csv('Marketing Customer Value Analysis DataSet.csv')
df.head()
```

```python
df.shape
```

## Scrubbing Data & Exploring Data

```python
df['Engaged'] = df['Response'].apply(lambda x : 0 if x == 'No' else 1)
df.head()
```

```python
list(df.columns)
```

```python
df.groupby('Engaged').count()
```

```python
df.groupby('Engaged').count()['Response']
```

```python
# engagement rate

engagement_rate_df = pd.DataFrame(df.groupby('Engaged').count()['Response']/df.shape[0]*100)
engagement_rate_df
```

```python
engagement_rate_df.T
```

```python
# renew offer

engagement_by_offer_type_df = pd.pivot_table(df, values='Response', index='Renew Offer Type',
                                            columns='Engaged', aggfunc=len).fillna(0.0)
engagement_by_offer_type_df.columns = ['Not Engaged', 'Engaged']
engagement_by_offer_type_df
```

```python
# plotting renew offer

engagement_by_offer_type_df.plot(kind='pie', figsize=(15, 7), startangle=90,
                                subplots=True, autopct=lambda x : '%0.1f%%' %x)
plt.show()
```

```python
# sales channel

engagement_by_sales_channel_df = pd.pivot_table(df, values='Response', index='Sales Channel',
                                               columns='Engaged', aggfunc=len).fillna(0.0)
engagement_by_sales_channel_df.columns = ['Not Engaged', 'Engaged']
engagement_by_sales_channel_df
```

```python
# plotting sales channel

engagement_by_sales_channel_df.plot(kind='pie', figsize=(15, 7), startangle=90,
                                subplots=True, autopct=lambda x : '%0.1f%%' %x)
plt.show()
```

```python
# total claim amount distributions by engagement(showfliers false)

ax = df[['Engaged', 'Total Claim Amount']].boxplot(by='Engaged', showfliers=False, figsize=(7, 5))
ax.set_xlabel('Engaged')
ax.set_ylabel('Total Claim Amount')
ax.set_title('Total Claim Amount Distributions by Engagements')
plt.suptitle("")
plt.show()
```

```python
# total claim amount distributions by engagement(showfliers true)

ax = df[['Engaged', 'Total Claim Amount']].boxplot(by='Engaged', showfliers=True, figsize=(7, 5))
ax.set_xlabel('Engaged')
ax.set_ylabel('Total Claim Amount')
ax.set_title('Total Claim Amount Distributions by Engagements')
plt.suptitle("")
plt.show()
```

```python
# income distributions by engagement(showfliers false)

ax = df[['Engaged', 'Income']].boxplot(by='Engaged', showfliers=False, figsize=(7, 5))
ax.set_xlabel('Engaged')
ax.set_ylabel('Income')
ax.set_title('Income Distributions by Engagements')
plt.suptitle("")
plt.show()
```

```python
# income distributions by engagement(showfliers true)

ax = df[['Engaged', 'Income']].boxplot(by='Engaged', showfliers=True, figsize=(7, 5))
ax.set_xlabel('Engaged')
ax.set_ylabel('Income')
ax.set_title('Income Distributions by Engagements')
plt.suptitle("")
plt.show()
```

## Modeling Data

```python
# continuous variables only modeling

continuous_vars = ['Customer Lifetime Value', 'Income', 'Monthly Premium Auto',
                  'Months Since Last Claim', 'Months Since Policy Inception',
                  'Number of Open Complaints', 'Number of Policies', 'Total Claim Amount']
```

```python
# logistic regression modeling

logit = sm.Logit(df['Engaged'], df[continuous_vars])
```

```python
# fitting

logit_fit = logit.fit()
```

```python
# logistic model summary

logit_fit.summary()
```

```python
# categorical variables only modeling

df.describe()
```

```python
# categorical variable series

categories = pd.Categorical(df['Education'], categories=['High School or Below', 'Bachelor', 'College', 'Master', 'Doctor'])
```

```python
df['EducationFactorized'] = categories.codes
```

```python
# add gender

gender_values, gender_labels = df['Gender'].factorize()
df['GenderFactorized'] = gender_values
```

```python
# logistic regression modeling w/categorical variables

logit = sm.Logit(df['Engaged'], df[['GenderFactorized', 'EducationFactorized']])
```

```python
# summary

logit_fit.summary()
```
