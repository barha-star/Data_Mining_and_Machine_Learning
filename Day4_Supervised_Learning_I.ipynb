# Supervised Learning I

```python
import pandas as pd
```

```python
# data load

df = pd.read_csv('data/retail_transaction_data.csv')
```

```python
# check th dataframe

df.head(10)
```

```python
# convert dtype of the InvoiceDate to datetime

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.head()
```

```python
# InvoiceDate Ascending Sort

df.sort_values('InvoiceDate')
```

```python
# create 'revenue' column

df['revenue'] = df['UnitPrice'] * df['Quantity']
df.head()
```

```python
# sort by invoice

operations = {'revenue':'sum', 'InvoiceDate':'first', 'CustomerID':'first'}
df = df.groupby('InvoiceNo').agg(operations)
df.head()
```

```python
# yearly sort

df['year'] = df['InvoiceDate'].apply(lambda x : x.year)
df.head()
```

```python
# from 2010 year

df['days_since'] = (pd.datetime(year=2010, month=12, day=31) - 
                   df['InvoiceDate']).apply(lambda x : x.days)
df.head()
```

```python
# sort by customerID in 2010

operations = {'revenue':'sum', 'days_since':['max', 'min', 'nunique']}
X = df[df['year'] == 2010].groupby('CustomerID').agg(operations)
X.head()
```

```python
X.columns.values
```

```python
# remove space bar

X.columns = [' '.join(col).strip() for col in X.columns.values]
X.head()
```

```python
# average order cost per day

X['average_order_cost'] = X['revenue sum'] / X['days_since nunique']
X.head()
```

```python
# sort by CustomerID in 2011

Y = df[df['year'] == 2011].groupby('CustomerID')['revenue'].sum()
Y
```

```python
# wrangled dataframe

wrangled_df = pd.concat([X, Y], axis=1)
wrangled_df.columns = ['2010 revenue', 'days_since_first_purchase',
                      'days_since_last_purchase', 'number_of_purchases',
                      'avg_order_cost', '2011 revenue']
wrangled_df.head(10)
```

```python
# null data handling

wrangled_df = wrangled_df[~wrangled_df['2010 revenue'].isnull()]
wrangled_df.head(10)
```

```python
wrangled_df.tail(10)
```

```python
# null data handling

wrangled_df = wrangled_df[~wrangled_df['2011 revenue'].isnull()]
wrangled_df.tail(10)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.figure(figsize=(10, 8))
plt.boxplot(wrangled_df['2011 revenue'])
plt.show()
```

```python
wrangled_df = wrangled_df[wrangled_df['2011 revenue'] < ((wrangled_df['2011 revenue'].median()) + wrangled_df['2011 revenue'].std()*3)]
```

```python
wrangled_df = wrangled_df[wrangled_df['2010 revenue'] < ((wrangled_df['2010 revenue'].median()) + wrangled_df['2010 revenue'].std()*3)]
```
